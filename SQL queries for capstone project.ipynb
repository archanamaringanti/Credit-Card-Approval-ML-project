```python
import pandas as pd

credit_data = pd.read_csv("credit_mergedata_new.csv")
credit_data.head()
```

```python
import duckdb

conn = duckdb.connect()
conn.register('credit_data', credit_data)
```

```python
conn.execute("select * from credit_data").fetchdf().head()
```

# 1. Group the customers based on their income type and find the average of        their annual income.

```python
conn.execute("SELECT Employment_status, ROUND(AVG(Annual_income), 2) AS average_annual_income FROM credit_data GROUP BY Employment_status ORDER BY average_annual_income DESC").fetchdf()
```

# 2. Find the female owners of cars and property.

```python
conn.execute("SELECT * FROM credit_data WHERE Gender = 'F' AND Has_car = 'Y' AND Has_property = 'Y'").fetchdf()
```

# 3. Find the male customers who are staying with their families.

```python
conn.execute("SELECT * FROM credit_data WHERE Gender ='M' AND Family_Members > 1").fetchdf()
```

```python
conn.execute("SELECT * FROM credit_data WHERE Gender ='M' AND Housing_type = 'With parents'").fetchdf()
```

# 4. Please list the top five people having the highest income.

Note: I have taken the cleaned dataset (after handling null values, renaming columns, extracting columns from existing columns) but not after outlier treatment.

```python
conn.execute("SELECT * FROM credit_data ORDER BY Annual_income DESC LIMIT 5").fetchdf()
```

# 5. How many married people are having bad credit?

```python
conn.execute("SELECT  COUNT(1) AS married_bad_credit_cnt FROM credit_data WHERE Marital_status IN ('Married', 'Civil marriage') AND Is_approved = 0").fetchdf()
```

# 6. What is the highest education level and what is the total count?

```python
conn.execute("SELECT Education, COUNT(1) AS people_cnt FROM credit_data GROUP BY Education").fetchdf()
```

Highest Education level is Academic Degree with a count of 2 followed by Higher Education with a count of 426.

```python
conn.execute("SELECT COUNT(1) AS people_cnt FROM credit_data WHERE Education = 'Academic degree'").fetchdf()
```

# 7. Between married males and females, who is having more bad credit? 

```python
conn.execute("SELECT Gender, COUNT(1) AS gender_bad_credit_cnt FROM credit_data WHERE Is_approved = 0 AND Marital_status IN ('Married', 'Civil marriage') GROUP BY Gender").fetchdf()
```

Married males are having more bad credit
